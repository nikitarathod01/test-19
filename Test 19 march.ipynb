# Task:2 Currency converter program.

Write a Currency converter program and allow the user to select the money exchange for different countries.

```python
# currency exchange rates as per today
exchange_rates = {
    'USD':1.0,
    'INR':74.52,
    'EUR':0.85,
    'GBP':0.73,
    'JPY':110.15,
    'CYR':7.19,
    'HKD':7.82
}

def convert_currency (amount, from_currency, to_currency):
    if from_currency not in exchange_rates or to_currency not in exchange_rates: 
        return "Currency not supported"
    
    converted_amount = amount*exchange_rates[to_currency]/exchange_rates[from_currency]
    return converted_amount

# Input from user
amount = float(input("Enter the amount to convert: "))
from_currency = input("Enter the currency to convert from (e.g., INR, EUR, GBP): ").upper()
to_currency = input("Enter the currency to convert from (e.g., INR, EUR, GBP): ").upper()

# Currency conversion
result = convert_currency(amount,from_currency,to_currency)
if result == "currency not supported":
    print("currency not supported please try again..")
else:
    print(f"{amount}{from_currency} is equal to {result:.2f}{to_currency}")
```
